```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_regresion_df=pd.read_csv('data_regression.csv')
```

```python
data_regresion_df
```

```python
x=data_regresion_df['X1']
y=data_regresion_df['y']
```

```python
x = x.values.reshape(-1, 1)
```

```python
data_regresion_model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
data_regresion_model.fit(x_train, y_train)
```

```python
y_real=-0.45*x + 5
plt.figure(figsize = (13,7))
plt.scatter(x,y, s = 2)
plt.plot(x, y_real, '--',label ='Curva Teórica', c = 'r')
plt.plot(x_test,data_regresion_model.predict(x_test),label ='Regresion Lineal', c = 'g')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = data_regresion_df[['X1','X2','X3']]
y = data_regresion_df['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
poly_features = PolynomialFeatures(degree=2)  # Crear instancia de PolynomialFeatures
X_train_poly = poly_features.fit_transform(X_train)  # Transformar características de entrenamiento
X_test_poly = poly_features.transform(X_test)  # Transformar características de prueba
```

```python
regression_model = LinearRegression()  # Crear instancia de LinearRegression
regression_model.fit(X_train_poly, y_train)  # Entrenar el modelo utilizando características polinómicas
```

```python
y_pred = regression_model.predict(X_test_poly)  # Realizar predicciones en el conjunto de prueba
```
