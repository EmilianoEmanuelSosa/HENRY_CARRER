### **1**

1. Defina dos listas de valores `x` e `y`, tal que los pares (x,y) cumplan una relación lineal del tipo $y = 2x + 1$. Defina la lista de valores de `x` con 100 puntos en el intervalo de valores que van desde -1 hasta 1.

```python
# Definir la lista de valores de x con 100 puntos en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Calcular la lista de valores de y utilizando la relación lineal y = 2x + 1
y = 2 * x + 1

# Graficar los puntos
plt.plot(x, y, 'o', label='(x, y)')

# Agregar etiquetas y título al gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relación lineal: y = 2x + 1')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

2. Grafique la curva generada por los valores (x,y) en color verde, con un ancho `linewidth = 3` y que sea una línea a trazos `linestyle = '--'`. Agregue al gráfico el título 'Relación Lineal'.

```python
# Definir la lista de valores de x con 100 puntos en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Calcular la lista de valores de y utilizando la relación lineal y = 2x + 1
y = 2 * x + 1

# Graficar los puntos
plt.plot(x, y, color='green', linewidth=3, linestyle='--', label='(x, y)')
# Agregar etiquetas y título al gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relación lineal')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

### **2**

Graficar tres subplots, con las curvas (x1,y1),(x2,y2) y (x3,y3), en una única figura. Apilar los subplots en forma vertical y de manera que compartan el eje x.

```python

# Generar puntos en el eje X
x = np.linspace(0, 2*np.pi, 100)

# Generar puntos en el eje Y utilizando diferentes funciones
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2 + np.sin(x)

# Crear los subgráficos y compartir los ejes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

# Trazar la curva en el primer subgráfico
ax1.plot(x, y1, color='blue')
ax1.set_ylabel('Función seno')

# Trazar la curva en el segundo subgráfico
ax2.plot(x, y2, color='red')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Función coseno')

# Trazar la curva en el tercer subgráfico
ax3.plot(x, y3, color='purple')
ax3.set_xlabel('Eje X')
ax3.set_ylabel('pi X pi')

# Ajustar los márgenes para evitar superposiciones
plt.tight_layout()

# Mostrar el gráfico
plt.show()


```
