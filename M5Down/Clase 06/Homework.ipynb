### **1**

1. Defina dos listas de valores `x` e `y`, tal que los pares (x,y) cumplan una relaci√≥n lineal del tipo $y = 2x + 1$. Defina la lista de valores de `x` con 100 puntos en el intervalo de valores que van desde -1 hasta 1.

```python
# Definir la lista de valores de x con 100 puntos en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Calcular la lista de valores de y utilizando la relaci√≥n lineal y = 2x + 1
y = 2 * x + 1

# Graficar los puntos
plt.plot(x, y, 'o', label='(x, y)')

# Agregar etiquetas y t√≠tulo al gr√°fico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relaci√≥n lineal: y = 2x + 1')

# Mostrar la leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

2. Grafique la curva generada por los valores (x,y) en color verde, con un ancho `linewidth = 3` y que sea una l√≠nea a trazos `linestyle = '--'`. Agregue al gr√°fico el t√≠tulo 'Relaci√≥n Lineal'.

```python
# Definir la lista de valores de x con 100 puntos en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Calcular la lista de valores de y utilizando la relaci√≥n lineal y = 2x + 1
y = 2 * x + 1

# Graficar los puntos
plt.plot(x, y, color='green', linewidth=3, linestyle='--', label='(x, y)')
# Agregar etiquetas y t√≠tulo al gr√°fico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relaci√≥n lineal')

# Mostrar la leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

### **2**

Graficar tres subplots, con las curvas (x1,y1),(x2,y2) y (x3,y3), en una √∫nica figura. Apilar los subplots en forma vertical y de manera que compartan el eje x.

```python

# Generar puntos en el eje X
x = np.linspace(0, 2*np.pi, 100)

# Generar puntos en el eje Y utilizando diferentes funciones
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2 + np.sin(x)

# Crear los subgr√°ficos y compartir los ejes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

# Trazar la curva en el primer subgr√°fico
ax1.plot(x, y1, color='blue')
ax1.set_ylabel('Funci√≥n seno')

# Trazar la curva en el segundo subgr√°fico
ax2.plot(x, y2, color='red')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Funci√≥n coseno')

# Trazar la curva en el tercer subgr√°fico
ax3.plot(x, y3, color='purple')
ax3.set_xlabel('Eje X')
ax3.set_ylabel('pi X pi')

# Ajustar los m√°rgenes para evitar superposiciones
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()


```
